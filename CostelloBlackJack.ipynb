```python
import random


ranks = ["Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace"]
suits = ["Spades", "Hearts", "Clubs", "Diamonds"]
values = {"Two": 2, "Three": 3, "Four": 4, "Five": 5, "Six":6, "Seven": 7,"Eight": 8, "Nine": 9, "Ten": 10, "Jack": 10, "Queen": 10, "King": 10, "Ace": 11}
class Card():
    
    #This class is to create the basics of a card
    
    def __init__(self, suit, rank):
        self.suit= suit
        self.rank = rank
        self.value = values[rank]
    
    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
class Deck():
    def __init__(self):
        self.thedeck = []
        
        #Goes through every combination of suit and rank, and creates a new deck each time it's called
        
        for rank in ranks:
            for suit in suits:
                self.thedeck.append(Card(suit, rank))
                
    def shufflethis(self):
        random.shuffle(self.thedeck)
        
    def letsseefirstcard(self):
        print(self.thedeck[0])
        
    def __str__(self):
        return f"this deck got {len(self.thedeck)} cards in it!"
    
```

```python
class Player():
    
    #This class is for attributes and actions associated with the Player each time he/she plays the game.
    #In our case, we are only tracking the balance of the Player.
    
    def __init__(self, balance):
        self.balance = balance
    
    def win(self, bet):
        self.balance = self.balance +bet
        print(f"\n+${bet:,} to your balance\nCurrent balance: ${self.balance:,}")
        return 
    
    def lose(self, bet):
        self.balance = self.balance - bet
        print(f"\n-${bet:,} \nCurrent Balance: ${self.balance:,}")
        return 
    def __str__(self):
        return f"you got ${self.balance:,}"
```

```python
def createalltotalpos(hand):
    #I LOVE this logic here. First we add the total normally and add the entry to the 
    #"total possibilities" array (totalpos). Then for every "Ace" there is in the deck, it looks at 
    #the previous total entry, removes 10 from it, and appends that value to the "totalpos" array
    
    totalpos=[]
    total=0
    
    for card in hand:
        total = card.value + total
    totalpos.append(total)
    
    for card in hand:
        if card.value == 11:
            totalpos.append(totalpos[len(totalpos)-1]-10)
            
    return totalpos

def checkbust(totalarraypos):
    #Checks if every potential total is busted and returns a bool. I could also use the min(totalarraypos) to check
    #I think this way is more unique
    
    bustcounter=0
    for t in totalarraypos:
        if t>21:  
            bustcounter=bustcounter+1 

    if bustcounter == len(totalarraypos):
        return True
    else: 
        return False
                
def checkblackjack(totalarraypos):
    #Checks if even one of the potential totals is the always elusive BlackJack
    
    for t in totalarraypos:
        if t == 21:
            print("\nYOU GOT A B L A C K J A C K !!\n")
            
            return True
    return False

def askhitorstand():
    #Purpose is to take user input in consideration for the next move in Black Jack
    #"checkblackjack" returns "True" is there is a Black Jack. Assuming there is, you automatically "Stand"
    #to see if the Dealer can match your Black Jack.
    
    sillyans = True
    if not checkblackjack(totalpos):
        while sillyans == True:   
            hitorstand = input("\nType H for hit or S for stand: ").lower()

            if hitorstand == "s" or hitorstand=='h':
                sillyans=False
            else:
                print("PLEASE TYPE H or S. We aren't leaving until you get this: ")
                sillyans=True
    else:
        hitorstand='s'
    return hitorstand
def comparehands(dealertotalpos,playertotalpos):
    #The "total possibilities" array (dealertotalpos and playertotalpos) starts with all aces as 11. 
    #A new entry is added each time there is an ace, removing 10 from the previous total entry.
    #Therefore the "total possibilities" arrays are aligned in descending order (highest value to lowest)
    #We want to find the highest total that isn't a bust of the Player and Dealer and compare them to see who wins
    
    playersbest =0
    dealersbest=0
    for currenttotal in playertotalpos:
        if currenttotal < 22:
            playersbest = currenttotal
            
            break
        else:
            continue
    for currenttotal in dealertotalpos:
        if currenttotal < 22:
            dealersbest = currenttotal
            break
        else:
            continue
   
    #From the way of finding the highest total of the dealer is set up, we know that if the "dealertotalpos" array
    #is exhausted and no new value is assigned to "dealersbest". All total possibilities must've been a bust.
    
    if dealersbest>playersbest:
        print(f"\n+Dealer's Best+: |{dealersbest}| > Your Best: |{playersbest}| You lose!")
        maincharacter.lose(bet)
    elif dealersbest==playersbest:
        print(f"\nDealer's Best: |{dealersbest}| = Your Best: |{playersbest}| No money lost!")
        print(f"Current Balance: ${maincharacter.balance:,}")
    else:
        if dealersbest == 0:
            dealersbest = "BUST"
        print(f"\nDealer's Best: |{dealersbest}| < +Your Best+: |{playersbest}| YOU WIN!!!")
        maincharacter.win(bet)
def gameend():
    #checks if you want to play again; then returns the "gameon" value
    print("\n^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
    while True:
        playagain = input("\nDo you wanna play again? Type y for yes or n for no: ").lower()
        if playagain == "y":
            gameon = True
            return gameon
        elif playagain == "n":
            print("Thanks for playing!")
            gameon = False
            return gameon
        else:
            print("TYPE N OR Y! STOP THIS MADNESS: ")
            continue
    
def printhands(totalarraypos, playershand):
    #Most people would not have done this part in their blackjack game. I thought it would be a fun challenge.
    
    #The ace makes displaying the value of the hands quite tricky. Know that "index" is the index of totalarraypos.
    #Know that "timesacechanged" is a counter of how many times we have changed the ace's value. The goal is one
    #per total to fit with the logic I had set.
    
    #The "total possibilities" array (totalarraypos) starts with all aces as 11. 
    #A new entry is added for each time there is an ace, removing 10 from the previous total entry.

    #Minus one from the index as we know the first time we want to change the ace is when the index is 1
    #We want to match the timesacechanged (starts at 0) variable to the index value. Below it's set up to change the
    #ace value once per iteration of displaying the hand
    
    handstatement = ""
    timesacechanged = 0
    for index, total in enumerate(totalarraypos):
        for card in playershand:
            if card.value==11 and timesacechanged == index-1:
                timesacechanged=timesacechanged +1
                card.value=card.value-10
        for card in playershand:       
            if card == playershand[0]:
                handstatement = handstatement + f'\nPlayer Hand {index+1}: {card.value}'
            else:
                handstatement = handstatement + f' + {card.value}'

        handstatement = handstatement + f' = |{total}|'
        if total > 21:
            handstatement = handstatement + " BUST!!!"
        elif total==21:
            handstatement = handstatement + " --> $+ Black Jack $+"
    print(handstatement)
    
    #At the end the card value of each ace is 1, so we have to reset it to 11 again
    
    for card in playershand:
        if card.value==1:
            card.value=11
```

```python
# Game
print("Welcome to Black Jack")

#To make it simple I only used integers and made sure the Player couldn't input a negative value
#You can't literally play with House money

while True:
    try:
        startingbalance = int(input("How much do you wanna risk today? "))
        if startingbalance < 0:
            raise ValueError
        else: break
    except ValueError:
        print("Please only enter positive counting numbers. Try again")
        
maincharacter = Player(startingbalance)
print(f"Now {maincharacter} to dance with. Let's get started")
```

```python
bet = 0
gameon = True
while gameon== True:
    
    print(f"Starting balance: $ {maincharacter.balance:,}")
    
    #I saved processing time here not using another while loop since the logic is at the top
    #In my game, the Player can end up owing $1,000 to the casino
    try:
        bet = int(input("Place your bet $"))
        if bet < 0:
            raise ValueError
    
    except ValueError:
        print("Please only enter positive counting numbers. Try again")
        continue
    
    if bet > maincharacter.balance + 1000:
        print("You have exceeded the max bet - $1,000 more than your remaining balance - please enter a lower amount")
        continue
    
    print("\n---------------------------------------------------------")
    playdeck = Deck()
    playdeck.shufflethis()
    
    playershand = []
    dealershand=[]

    for x in range(0,2):
        currentcard = playdeck.thedeck.pop(0)
        playershand.append(currentcard)
        dealercard = playdeck.thedeck.pop(0)
        dealershand.append(dealercard)
    
    totalpos = createalltotalpos(playershand)
    
    
    print(f"\nDealer has a {dealershand[0]} and a hidden card.\nDealer: |{dealershand[0].value}| up")

    print(f"\nYou have a {playershand[0]} and a {playershand[1]}")

    printhands(totalpos,playershand)

    #Asks for the user's input for the next move when applicable
    hitorstand=askhitorstand()

    while hitorstand=="h":
        
        andhand= "\nYou have"
        
        currentcard = playdeck.thedeck.pop(0)
        playershand.append(currentcard)
        print(f"\nHIT! Added {currentcard}")
        
        #Here we format the cards to work well with English grammar 
        for card in playershand:
            if card == playershand[0]:
                andhand= andhand + f" a {card}"
            elif x != playershand[-1]:
                andhand= andhand + f", a {card}"
            else:
                andhand= andhand + f", and a {card}"
                
        totalpos = createalltotalpos(playershand)

        print(andhand)
        printhands(totalpos,playershand)
        
        #Check bust returns True if all possible hands are busts
        if checkbust(totalpos):
            print("\nCompletely. BUSTED.")
            maincharacter.lose(bet)
            gameon = gameend()
            break

        hitorstand=askhitorstand()
        
    if hitorstand == "s":
        print("\n---------------------------------------------------------")
        dealertotalpos = createalltotalpos(dealershand)
        
        print(f"\nDealer has {dealershand[0]} and {dealershand[1]}")
        
        #I wanted to show two different ways of displaying hands. This one is much simpler
        for index,tot in enumerate(dealertotalpos):
            print(f'Dealer Hand {index+1}: |{tot}|')
        
        #Out of all possibile two card totals there is only one case where the max total 
        #of the dealershand is over 17, and the dealer still needs to hit. That is when the dealer has two aces.
        #"Repeat" is for when the dealer does "Hit" and needs to hit more
        
        repeat = False
        while max(dealertotalpos) <17 or min(dealertotalpos) == 2 or repeat:
            repeat = False
            dealerstring=''
            
            dealercard = playdeck.thedeck.pop(0)
            dealershand.append(dealercard)
            
            dealertotalpos = createalltotalpos(dealershand)
            
            print(f"\nDealer HITS! Added {dealershand[-1]}")
            
            for index,tot in enumerate(dealertotalpos):
                print(f'AFTER HIT! Dealer Hand {index+1}: |{tot}|')
            
            #Goes through every "total possibility" (dealertotalpos) and sees if the dealer needs to stop hitting
            #Dealer needs to stop hitting if they have a total between 21 and 17, or if all hands are busted
            for total in dealertotalpos:
                if total <= 21 and tot >=17 or min(dealertotalpos)>21:
                    repeat = False
                    break
                else:
                    repeat=True
        
        #Compare hands puts the dealers best total versus yours and gives a winner
        comparehands(dealertotalpos,totalpos)
        
        gameon = gameend()
```
